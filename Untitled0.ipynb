b.	Python code for file handling

```python
# File Handling Example
filename = "example.txt"

# Writing to a file
with open(filename, "w") as file:
    file.write("This is a sample file for cloud computing lab.")

# Reading from a file
with open(filename, "r") as file:
    content = file.read()

print("File content:", content)
```

c.	Python code for image processing

```python
# Image Processing Example using OpenCV
import cv2
from google.colab.patches import cv2_imshow

# Load an image (Upload one first using the upload cell below)
from google.colab import files
uploaded = files.upload()

for fn in uploaded.keys():
    img = cv2.imread(fn)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2_imshow(gray)
```

d.	Python code for handling video files

```python
import cv2
from google.colab import files

# Upload a video file
uploaded = files.upload()

for fn in uploaded.keys():
    cap = cv2.VideoCapture(fn)
    if not cap.isOpened():
        print("Error opening video file")
    else:
        print(f"Video {fn} loaded successfully!")

cap.release()
```
